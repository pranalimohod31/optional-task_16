```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('/content/fifa_world_cup_2026_player_performance.csv')
```

```python
print(df.head(10))
```

```python
print(df.shape)
```

```python
df.info()
```

```python
print(df.describe())
```

```python
print(df.isnull().sum())
```

```python
num_cols = df.select_dtypes(include=['int64','float64']).columns
cat_cols = df.select_dtypes(include=['object']).columns

print(num_cols)
print(cat_cols)
```

```python
sns.histplot(df['age'], kde=True)
plt.title('Age Distribution')
plt.show()
```

```python
sns.countplot(x='position', data=df)
plt.title('Player Positions')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(df.corr(numeric_only=True), cmap='coolwarm')
plt.show()
```

```python
sns.boxplot(x='preferred_foot', y='player_rating', data=df)
plt.title('Player Rating by Preferred Foot')
plt.show()
```

Summary

- The dataset contains player performance data from the FIFA World Cup 2026.
- It includes player details, match statistics, physical attributes, and performance scores.
- Missing values (if any) were identified using `isnull()`.
- The distribution plots show how player ages are spread.
- Count plots show the number of players in each playing position.
- The heatmap helps identify relationships between numerical features.
- Boxplots compare player ratings based on their preferred foot.
